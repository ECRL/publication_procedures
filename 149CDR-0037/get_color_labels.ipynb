```python
COLORS = [
    'tab:blue',
    'tab:orange',
    'tab:green',
    'tab:red',
    'tab:purple',
    'tab:brown',
    'tab:pink',
    'tab:gray',
    'tab:olive',
    'tab:cyan'
]
```

```python
from matplotlib import pyplot as plt

for i in range(len(COLORS)):
    plt.scatter([i + 3], 5, color=COLORS[i], s=[500], marker='*')
plt.show()
```
