```python
from ecnet.datasets import load_mon, load_ron

dataset_mon = load_mon(as_dataset=True, backend='alvadesc')
dataset_ron = load_ron(as_dataset=True, backend='alvadesc')
```

```python
from ecnet.tasks.feature_selection import select_rfr

desc_idx_mon, desc_imp_mon = select_rfr(dataset_mon, total_importance=1.0, n_estimators=50, n_jobs=8)
desc_idx_ron, desc_imp_ron = select_rfr(dataset_ron, total_importance=1.0, n_estimators=50, n_jobs=8)
dataset_mon.set_desc_index(desc_idx_mon)
dataset_ron.set_desc_index(desc_idx_ron)
```

```python
print(dataset_mon.desc_names[0], desc_imp_mon[0])
print(dataset_ron.desc_names[0], desc_imp_ron[0])
```

```python
from matplotlib import pyplot as plt

plt.clf()
experimental_mon = [tv[0] for tv in dataset_mon.target_vals]
plt.scatter([c[0] for c in dataset_mon.desc_vals], experimental_mon, color='blue')
plt.xlabel('{} Value'.format(dataset_mon.desc_names[0]))
plt.ylabel('Experimental MON')
plt.show()
```

```python
plt.clf()
experimental_ron = [tv[0] for tv in dataset_ron.target_vals]
plt.scatter([c[0] for c in dataset_ron.desc_vals], experimental_ron, color='blue')
plt.xlabel('{} Value'.format(dataset_ron.desc_names[0]))
plt.ylabel('Experimental RON')
plt.show()
```
