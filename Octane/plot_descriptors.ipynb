```python
# Open training, testing subsets from https://doi.org/10.1021/acs.energyfuels.0c01533
from csv import DictReader

with open('octane_database.csv', 'r') as csv_file:
    reader = DictReader(csv_file)
    compounds = [r for r in reader]
csv_file.close()
```

```python
# Format data
from alvadescpy import smiles_to_descriptors

smiles = [c['canonical_smiles'] for c in compounds]
mon = [float(c['properties.motor_octane_number.value']) for c in compounds]
ron = [float(c['properties.research_octane_number.value']) for c in compounds]
os = [ron[i] - mon[i] for i in range(len(compounds))]

desc_vals = [smiles_to_descriptors(smi) for smi in smiles]
```

```python
# GATS2m plotting
from matplotlib import pyplot as plt

gats2m = [d['GATS2m'] for d in desc_vals]

plt.clf()
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams.update({'font.size': 12})
plt.scatter(gats2m, ron, color=(0, 114/235, 178/235), alpha=0.5)
plt.xlabel('GATS2m Value')
plt.ylabel('RON Value')
plt.savefig('plots/GATS2m_RON.png', dpi=256)
plt.show()

plt.clf()
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams.update({'font.size': 12})
plt.scatter(gats2m, mon, color=(0, 158/235, 115/235), alpha=0.5)
plt.xlabel('GATS2m Value')
plt.ylabel('MON Value')
plt.savefig('plots/GATS2m_MON.png', dpi=256)
plt.show()

plt.clf()
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams.update({'font.size': 12})
plt.scatter(gats2m, os, color=(213/235, 94/235, 0), alpha=0.5)
plt.xlabel('GATS2m Value')
plt.ylabel('OS Value')
plt.savefig('plots/GATS2m_OS.png', dpi=256)
plt.show()
```

```python
plt.clf()
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams.update({'font.size': 12})
fig, ax = plt.subplots()
ax.set_ylabel('OS Value')
ax.set_xlabel('nCsp2 Value')
pos = [0, 1, 2, 4, 6, 8]
nc0 = [os[idx] for idx, nc in enumerate(ncsp2) if nc == 0]
nc1 = [os[idx] for idx, nc in enumerate(ncsp2) if nc == 1]
nc2 = [os[idx] for idx, nc in enumerate(ncsp2) if nc == 2]
nc4 = [os[idx] for idx, nc in enumerate(ncsp2) if nc == 4]
nc6 = [os[idx] for idx, nc in enumerate(ncsp2) if nc == 6]
nc8 = [os[idx] for idx, nc in enumerate(ncsp2) if nc == 8]
nc_stats = [nc0, nc1, nc2, nc4, nc6, nc8]
parts = ax.violinplot(nc_stats, pos, showmeans=True, showextrema=True)
for pt in parts['bodies']:
    pt.set_facecolor((213/235, 94/235, 0))
    pt.set_alpha(0.5)
for partname in ('cbars','cmins','cmaxes','cmeans'):
    parts[partname].set_edgecolor((213/235, 94/235, 0))
plt.savefig('plots/nCsp2_violin.png', dpi=256)
plt.show()
```
