```python
from ecnet.datasets import load_ysi

smiles_ysi, ysi = load_ysi()
ysi = [i[0] for i in ysi]
```

```python
from matplotlib import pyplot as plt

plt.clf()
plt.rcParams['font.family'] = 'Times New Roman'
plt.hist(ysi, color=(0, 114/235, 178/235), bins=15, edgecolor='black', linewidth=1.1)
plt.xlabel('Experimental YSI')
plt.ylabel('Number of Compounds')
plt.show()
```

```python
ysi_crop = [i for i in ysi if i < 300]

print(len(ysi_crop) / len(ysi))
```

```python
plt.clf()
plt.rcParams['font.family'] = 'Times New Roman'
plt.hist(ysi_crop, color=(0, 114/235, 178/235), bins=15, edgecolor='black', linewidth=1.1)
plt.xlabel('Experimental YSI')
plt.ylabel('Number of Compounds')
plt.show()
```

```python
ysi_sub_100 = [i for i in ysi_crop if i < 150]

print(len(ysi_sub_100) / len(ysi_crop))
print(len(ysi_sub_100) / len(ysi))
```
