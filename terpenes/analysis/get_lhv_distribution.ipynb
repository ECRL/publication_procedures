```python
from ecnet.datasets import load_lhv

PROP_SHORTHAND = "LHV (MJ/kg)"
PROP_ABVR = "lhv"

_, compounds = load_lhv()
compounds = [i[0] for i in compounds]
print(len(compounds))
```

```python
from matplotlib import pyplot as plt

plt.clf()
plt.rcParams["font.family"] = "Times New Roman"
fig = plt.figure(figsize=(3.75, 3.75))
ax = fig.subplots()
ax.hist(compounds, color=(0, 114/235, 178/235), bins=15, edgecolor='black', linewidth=1.1)
ax.set_xlabel(f"Experimental {PROP_SHORTHAND}")
ax.set_ylabel("Number of compounds")
plt.savefig(f"figures/distr_{PROP_ABVR}.png", dpi=256)
plt.show()
```
