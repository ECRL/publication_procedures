```python
from csv import DictReader

PROPERTY = "properties.research_octane_number.value"
PROP_SHORTHAND = "RON"
PROP_ABVR = "ron"

with open("data/combustdb_master.csv", "r") as csv_file:
    reader = DictReader(csv_file)
    rows = [r for r in reader]
csv_file.close()

compounds = [float(r[PROPERTY]) for r in rows if r[PROPERTY] != "-"]
print(len(compounds))
```

```python
from matplotlib import pyplot as plt

plt.clf()
plt.rcParams["font.family"] = "Times New Roman"
plt.hist(compounds, color=(0, 114/235, 178/235), bins=15, edgecolor='black', linewidth=1.1)
plt.xlabel(f"Experimental {PROP_SHORTHAND}")
plt.ylabel("Number of compounds")
plt.savefig(f"figures/distr_{PROP_ABVR}.png", dpi=256)
plt.show()
```
