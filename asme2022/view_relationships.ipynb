```python
from ecnet.datasets import load_cn, load_ysi, load_lhv

smiles_cn, cn = load_cn()
smiles_ysi, ysi = load_ysi()
smiles_lhv, lhv = load_lhv()

smiles_to_use = []
cn_to_use = []
ysi_to_use = []
lhv_to_use = []

for idx_cn, smi_cn in enumerate(smiles_cn):
    has_ysi = False
    ysi_idx = None
    has_lhv = False
    lhv_idx = None
    for idx, smi in enumerate(smiles_ysi):
        if smi_cn == smi:
            has_ysi = True
            ysi_idx = idx
            break
    for idx, smi in enumerate(smiles_lhv):
        if smi_cn == smi:
            has_lhv = True
            lhv_idx = idx
            break
    if has_ysi and has_lhv:
        smiles_to_use.append(smi_cn)
        cn_to_use.append(cn[idx_cn])
        ysi_to_use.append(ysi[ysi_idx])
        lhv_to_use.append(lhv[lhv_idx])

print(len(smiles_to_use), len(cn_to_use), len(ysi_to_use), len(lhv_to_use))
```

```python
from matplotlib import pyplot as plt

plt.clf()
plt.rcParams['font.family'] = 'Times New Roman'
plt.xlabel('CN')
plt.ylabel('YSI')
plt.scatter(cn_to_use, ysi_to_use, color='blue', alpha=0.5)
plt.show()
```

```python
plt.clf()
plt.rcParams['font.family'] = 'Times New Roman'
plt.xlabel('CN')
plt.ylabel('LHV')
plt.scatter(cn_to_use, lhv_to_use, color='blue', alpha=0.5)
plt.show()
```

```python
plt.clf()
plt.rcParams['font.family'] = 'Times New Roman'
plt.xlabel('LHV')
plt.ylabel('YSI')
plt.scatter(lhv_to_use, ysi_to_use, color='blue', alpha=0.5)
plt.show()
```
